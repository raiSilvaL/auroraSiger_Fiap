###1.1 Organização e descrição de telemetria
---
A telemetria é um processo automatizado de coleta e transmissão de dados de medições ou outras informações de pontos remotos para equipamentos receptores para monitoramento e análise 1. O termo deriva do grego tele (distância) e metron (medida), refletindo sua essência de medição à distância 2.

**Funcionamento da Telemetria**

O sistema de telemetria opera em etapas fundamentais:

1.   Captura de Dados: Sensores instalados em equipamentos ou ambientes remotos coletam dados físicos (temperatura, pressão, nível de energia, etc.) ou de status (integridade estrutural, funcionamento de módulos).
2.   Transmissão: Os dados coletados são transmitidos automaticamente para uma estação receptora. Essa transmissão pode ocorrer via ondas de rádio, satélite, redes celulares ou outras tecnologias de comunicação, dependendo da aplicação e da distância.
3.   Recepção e Processamento: Na estação receptora, os dados são recebidos, decodificados e processados. Isso pode envolver armazenamento, visualização em tempo real e análise para identificar tendências, anomalias ou condições críticas.

**Dados da temetria da Aurora Siger**

* Temperatura interna: Entre -10°C e 40;
* Temperatura externa: Entre -120°C e 120°C;
* Integridade estrutural: 0 Com falhas e 1 Intacta;
* Níveis de energia: Maior que 95%;
* Pressão dos tanques: Entre 700.000 Pa e 1.200.000 Pa;
* Status dos módulos críticos (Navegação, Comunicação, Controle de altitude, Propulsão e Suporte de Vida): 0 Com falha ou 1 Operacional
