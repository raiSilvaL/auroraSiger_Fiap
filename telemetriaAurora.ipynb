###1.1 Organização e Descrição da Telemetria
---

A telemetria é um processo automatizado de coleta e transmissão de dados de medições ou outras informações de pontos remotos para equipamentos receptores para monitoramento e análise. O termo deriva do grego tele (distância) e metron (medida), refletindo sua essência de medição à distância.

**Funcionamento da Telemetria**

O sistema de telemetria opera em etapas fundamentais:

* **Captura de Dados:** Sensores instalados em equipamentos ou ambientes remotos coletam dados físicos (temperatura, pressão, nível de energia, etc.) ou de status (integridade estrutural, funcionamento de módulos).

* **Transmissão:** Os dados coletados são transmitidos automaticamente para uma estação receptora. Essa transmissão pode ocorrer via ondas de rádio, satélite, redes celulares ou outras tecnologias de comunicação, dependendo da aplicação e da distância.

* **Recepção e Processamento:** Na estação receptora, os dados são recebidos, decodificados e processados. Isso pode envolver armazenamento, visualização em tempo real e análise para identificar tendências, anomalias ou condições críticas.

**Dados da Telemetria da Aurora Siger**

Para a monitorização da aeronave Aurora Siger, os dados de telemetria são cruciais para garantir a segurança e a operacionalidade. A seguir, são detalhados os parâmetros a serem monitorizados e suas faixas de operação:

| Parâmetro | Unidade/Formato | Faixa de Operação | Descrição |
| --- | --- | --- | --- |
| Temperatura Interna | °C | Entre -10°C e 40°C | Temperatura dentro da cabine ou compartimento crítico. |

###1.2 Algoritimo de verificação
---

O algoritmo de verificação de decolagem tem como objetivo determinar se a aeronave está em condições seguras para decolar, com base nos dados de telemetria e nas faixas seguras predefinidas na seção 1.1.

  **Pseudocódigo do Algoritmo**

```
  FUNÇÃO VerificarDecolagem(
    tempInterna, tempExterna, integEstrutural, nivelEnergia, pressaoTanque,
    moduNavegação, moduComunicação, moduAltitude, moduPropulsao, moduSupVida
):
    SE tempInterna >= -10 E tempInterna <= 40 E
       tempExterna >= -120 E tempExterna <= 120 E
       integEstrutural = 1 E
       nivelEnergia > 95 E
       pressaoTanque >= 700000 E pressaoTanque <= 1200000 E
       moduNavegação = 1 E
       moduComunicação = 1 E
       moduAltitude = 1 E
       moduPropulsao = 1 E
       moduSupVida = 1 ENTAO
        RETORNAR "PRONTO PARA DECOLAR"
    SENÃO
        RETORNAR "DECOLAGEM ABORTADA"
FIM FUNÇÃO
```

###1.3 Script em python
---

Esta seção detalhará a implementação do algoritmo de verificação de decolagem em Python. O script simulará a leitura dos dados de telemetria, executará as verificações definidas no algoritmo e imprimirá o resultado final. A estrutura básica do script será a seguinte:

1. **Simulação de Leitura de Dados:** Geração de dados de telemetria aleatórios.

2. **Função de Verificação:** Implementação da lógica do pseudocódigo.

3. **Execução e Impressão:** Chamada da função de verificação e exibição do resultado.



````python
import random

def verificar_decolagem (tempInterna, tempExterna, integEstrutural, nivelEnergia, pressaoTanque,
    moduNavegação, moduComunicação, moduAltitude, moduPropulsao, moduSupVida):

  if (tempInterna >= -10 and tempInterna <= 40 and
      tempExterna >= -120 and tempExterna <= 120 and
      integEstrutural == 1 and
      nivelEnergia > 95 and
      pressaoTanque >= 700000 and pressaoTanque <= 1200000 and
      moduNavegação == 1 and
      moduComunicação == 1 and
      moduAltitude == 1 and
      moduPropulsao == 1 and
      moduSupVida == 1
    ):
      return "PRONTO PARA DECOLAR"
  else:
      return "DECOLAGEM ABORTADA"

#Simulador de decolagem (Mudar faixas para tester)
def simular_telemetria():
    temperatura_interna = round(random.uniform(-10, 40),2) # Faixa: -10 a 40
    temperatura_externa = round(random.uniform(-120, 120),2) # Faixa: -120 a 120
    integridade_estrutural = random.choice([1]) # 1 para OK
    niveis_energia = round(random.uniform(95, 100),2) # Faixa: > 95
    pressao_tanques = round(random.uniform(700000, 1200000),0) # Faixa: 700k a 1.2M

    status_modulos = [random.choice([1]) for _ in range(5)] # 1 para OK
    moduNavegação, moduComunicação, moduAltitude, moduPropulsao, moduSupVida = status_modulos

    return {
      "tempInterna": temperatura_interna,
      "tempExterna": temperatura_externa,
      "integEstrutural": integridade_estrutural,
      "nivelEnergia": niveis_energia,
      "pressaoTanque": pressao_tanques,
      "moduNavegação": moduNavegação,
      "moduComunicação": moduComunicação,
      "moduAltitude": moduAltitude,
      "moduPropulsao": moduPropulsao,
      "moduSupVida": moduSupVida
    }

if __name__ == "__main__":
  print("Simulando dados de telemetria e verificando decolagem...")

  # Teste com dados simulados
  dados_telemetria = simular_telemetria()
  print(f"\nDados de Telemetria: {dados_telemetria}")
  resultado_simulado = verificar_decolagem(**dados_telemetria)
  print(f"Resultado da Verificação (Simulado): {resultado_simulado}")
````

###1.4 Análise energética
---

A análise energética é fundamental para determinar a viabilidade de uma decolagem e a autonomia inicial da aeronave. Para esta análise, consideraremos os seguintes parâmetros hipotéticos para a aeronave Aurora Siger:

* Capacidade Total da Bateria: 500 kWh
* Carga Atual da Bateria: 98%
* Consumo Estimado na Decolagem: 150 kWh
* Perdas Energéticas (durante a decolagem e standby): 5% da carga atual

**Cálculo da Autonomia Inicial**

1. Carga Atual Disponível (kWh):
```
  Carga Atual Disponível = Capacidade Total * Carga Atual (%)
  Carga Atual Disponível = 500 kWh * 0.98 = 490 kWh
```

2. Perdas Energéticas (kWh):
```
  Perdas Energéticas = Carga Atual Disponível * Perdas Energéticas (%)
  Perdas Energéticas = 490 kWh * 0.05 = 24.5 kWh
```

3. Energia Líquida Disponível para Decolagem (kWh):
```
  Energia Líquida Disponível = Carga Atual Disponível - Perdas Energéticas
  Energia Líquida Disponível = 490 kWh - 24.5 kWh = 465.5 kWh
```

4. Energia Remanescente Após Decolagem (kWh):
```
  Energia Remanescente = Energia Líquida Disponível - Consumo Estimado na Decolagem
  Energia Remanescente = 465.5 kWh - 150 kWh = 315.5 kWh
```

**Conclusão da Análise Energética**

Com base nos cálculos, a aeronave Aurora Siger teria 315.5 kWh de energia remanescente após a decolagem, considerando a capacidade total, carga atual, consumo estimado e perdas energéticas. Este valor indica a energia disponível para as fases subsequentes do voo, como cruzeiro e pouso. É crucial que a energia remanescente seja suficiente para a missão planejada, além de uma reserva de segurança
